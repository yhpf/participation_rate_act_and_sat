##### 12. Write a function using only list comprehensions, no loops, to compute standard deviation. Using this function, calculate the standard deviation of each numeric column in both data sets. Add these to a list called `sd`.

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
import numpy as np 

# Here define a function (std) that calculates the standard deviation, using the given formula.
def std(series):
    mean = np.mean(series)
    squares = [(x-mean)**2 for x in series]
    sum_square = np.sum(squares)
    stdev = np.sqrt((1.0/len(series))* sum_square)
    return stdev

# I use the function above (where I calculate the standard deviation) to look at all column in test_data.
# However the first column is states and statenames are not numbers, it's strings. 
# Therefore i use .loc to tell that I want to use all rows but only column from act_part to sat_math. 
# Excluding the first column states.
sd = test_data.apply(std)
sd
```
